#List Nama yang available PPS



```python
bisa_senin = []

for i in range(0, len(df_jkantor.NAMA_PEGAWAI)):
  if df_jkantor["SENIN"][i] == 'WFO':
    if df_jkantor['NAMA_PEGAWAI'][i] not in SENIN:
      bisa_senin.append(df_jkantor['NAMA_PEGAWAI'][i])
```

```python
bisa_selasa = []

for i in range(0, len(df_jkantor.NAMA_PEGAWAI)):
  if df_jkantor["SELASA"][i] == 'WFO':
    if df_jkantor['NAMA_PEGAWAI'][i] not in SELASA:
      bisa_selasa.append(df_jkantor['NAMA_PEGAWAI'][i])
```

```python
bisa_rabu = []

for i in range(0, len(df_jkantor.NAMA_PEGAWAI)):
  if df_jkantor["RABU"][i] == 'WFO':
    if df_jkantor['NAMA_PEGAWAI'][i] not in RABU:
      bisa_rabu.append(df_jkantor['NAMA_PEGAWAI'][i])
```

```python
bisa_kamis = []

for i in range(0, len(df_jkantor.NAMA_PEGAWAI)):
  if df_jkantor["KAMIS"][i] == 'WFO':
    if df_jkantor['NAMA_PEGAWAI'][i] not in KAMIS:
      bisa_kamis.append(df_jkantor['NAMA_PEGAWAI'][i])
```

```python
bisa_jumat = []

for i in range(0, len(df_jkantor.NAMA_PEGAWAI)):
  if df_jkantor["JUMAT"][i] == 'WFO':
    if df_jkantor['NAMA_PEGAWAI'][i] not in JUMAT:
      bisa_jumat.append(df_jkantor['NAMA_PEGAWAI'][i])
```

```python
data = {'Senin': pd.Series(bisa_senin),
        'Selasa': pd.Series(bisa_selasa),
        'Rabu': pd.Series(bisa_rabu),
        'Kamis': pd.Series(bisa_kamis),
        'Jumat': pd.Series(bisa_jumat)}

bisa_piket = pd.DataFrame(data)
bisa_piket
```

```python
bisa_piket.to_excel('/content/day_available.xlsx')
```

```python
jadwal_senin = []
jadwal_selasa = []
jadwal_rabu = []
jadwal_kamis = []
jadwal_jumat = []

batas = 14

for nama in df_jkantor.NAMA_PEGAWAI.values:
  if nama in bisa_piket.Senin.values:
    if len(jadwal_senin) < batas:
      jadwal_senin.append(nama)
    else: break
  
for nama in df_jkantor.NAMA_PEGAWAI.values:
  if nama in bisa_piket.Selasa.values:
    if nama not in jadwal_senin:
      if len(jadwal_selasa) < batas:
        jadwal_selasa.append(nama)
      else: break

for nama in df_jkantor.NAMA_PEGAWAI.values:
  if nama in bisa_piket.Rabu.values:
    if nama not in jadwal_senin + jadwal_selasa:
      if len(jadwal_rabu) < batas:
        jadwal_rabu.append(nama)
      else: break

for nama in df_jkantor.NAMA_PEGAWAI.values:
  if nama in bisa_piket.Kamis.values:
    if nama not in jadwal_senin + jadwal_selasa + jadwal_rabu:
      if len(jadwal_kamis) < batas:
        jadwal_kamis.append(nama)
      else: break

for nama in df_jkantor.NAMA_PEGAWAI.values:
  if nama in bisa_piket.Jumat.values:
    if nama not in jadwal_senin + jadwal_selasa + jadwal_rabu + jadwal_kamis:
      if len(jadwal_jumat) < batas:
        jadwal_jumat.append(nama)
      else: break
```

```python
data2 = {'Senin': pd.Series(jadwal_senin),
        'Selasa': pd.Series(jadwal_selasa),
        'Rabu': pd.Series(jadwal_rabu),
        'Kamis': pd.Series(jadwal_kamis),
        'Jumat': pd.Series(jadwal_jumat)}

jadwal_piket = pd.DataFrame(data2)
jadwal_piket
```

```python
jadwal_piket.to_excel('/content/jadwal_piket.xlsx')
```

# Koordinator

```python
co = ['M. Rizal Gunawan','Irwansyah Lubis','Mufaidi','Djoko Prakosa','Soelarso','Arif Wibawa','Sonny','M. Idham','Achmad Supriyanto','Zia Ul Haq']
```

```python
kpp = df_jkantor[df_jkantor["Unit_Organisasi"]=="KPP Pratama"]["NAMA_PEGAWAI"]
kpp = pd.DataFrame(kpp)
kpp = kpp[~kpp.NAMA_PEGAWAI.isin(co)]
kpp.reset_index(drop=True, inplace=True)
kpp
```

```python
pel = df_jkantor[df_jkantor["Unit_Organisasi"]=="Seksi Pelayanan"]["NAMA_PEGAWAI"]
pel = pd.DataFrame(pel)
pel.reset_index(drop=True, inplace=True)
pel
```

```python
co_senin = []
co_selasa = []
co_rabu = []
co_kamis = []
co_jumat = []

kpp_senin = []
kpp_selasa = []
kpp_rabu = []
kpp_kamis = []
kpp_jumat = []

pel_senin = []
pel_selasa = []
pel_rabu = []
pel_kamis = []
pel_jumat = []

jadwal_senin = []
jadwal_selasa = []
jadwal_rabu = []
jadwal_kamis = []
jadwal_jumat = []

batas = 6 
batas_co = 2
batas_pel = 1
batas_kpp = 1

#senin
for i in range(0, len(df_jkantor.NAMA_PEGAWAI)): 
  if df_jkantor["NAMA_PEGAWAI"][i] in bisa_piket.Senin.values:
    if df_jkantor["NAMA_PEGAWAI"][i] in co:
      if len(co_senin) < batas_co:
        co_senin.append(df_jkantor["NAMA_PEGAWAI"][i])
      
    elif df_jkantor["NAMA_PEGAWAI"][i] in kpp.values:
      if len(kpp_senin) < batas_kpp:
        kpp_senin.append(df_jkantor["NAMA_PEGAWAI"][i])
      
    elif df_jkantor["NAMA_PEGAWAI"][i] in pel.values:
      if len(pel_senin) < batas_pel:
        pel_senin.append(df_jkantor["NAMA_PEGAWAI"][i])
     
    else: 
      if len(jadwal_senin) < batas:
        jadwal_senin.append(df_jkantor["NAMA_PEGAWAI"][i])
      

#selasa
for i in range(0, len(df_jkantor.NAMA_PEGAWAI)): 
  if df_jkantor["NAMA_PEGAWAI"][i] in bisa_piket.Selasa.values:
    if df_jkantor["NAMA_PEGAWAI"][i] in co:
      if df_jkantor["NAMA_PEGAWAI"][i] not in co_senin:
        if len(co_selasa) < batas_co:
          co_selasa.append(df_jkantor["NAMA_PEGAWAI"][i])
        
    elif df_jkantor["NAMA_PEGAWAI"][i] in kpp.values:
      if df_jkantor["NAMA_PEGAWAI"][i] not in kpp_senin:
        if len(kpp_selasa) < batas_kpp:
          kpp_selasa.append(df_jkantor["NAMA_PEGAWAI"][i])
        
    elif df_jkantor["NAMA_PEGAWAI"][i] in pel.values:
      if df_jkantor["NAMA_PEGAWAI"][i] not in pel_senin:
        if len(pel_selasa) < batas_pel:
          pel_selasa.append(df_jkantor["NAMA_PEGAWAI"][i])
        
    else: 
      if df_jkantor["NAMA_PEGAWAI"][i] not in jadwal_senin:
        if len(jadwal_selasa) < batas:
          jadwal_selasa.append(df_jkantor["NAMA_PEGAWAI"][i])
        

#rabu
for i in range(0, len(df_jkantor.NAMA_PEGAWAI)): 
  if df_jkantor["NAMA_PEGAWAI"][i] in bisa_piket.Rabu.values:
    if df_jkantor["NAMA_PEGAWAI"][i] in co:
      if df_jkantor["NAMA_PEGAWAI"][i] not in co_senin + co_selasa:
        if len(co_rabu) < batas_co:
          co_rabu.append(df_jkantor["NAMA_PEGAWAI"][i])
        
    elif df_jkantor["NAMA_PEGAWAI"][i] in kpp.values:
      if df_jkantor["NAMA_PEGAWAI"][i] not in kpp_senin + kpp_selasa:
        if len(kpp_rabu) < batas_kpp:
          kpp_rabu.append(df_jkantor["NAMA_PEGAWAI"][i])
        
    elif df_jkantor["NAMA_PEGAWAI"][i] in pel.values:
      if df_jkantor["NAMA_PEGAWAI"][i] not in pel_senin + pel_selasa:
        if len(pel_rabu) < batas_pel:
          pel_rabu.append(df_jkantor["NAMA_PEGAWAI"][i])
        
    else: 
      if df_jkantor["NAMA_PEGAWAI"][i] not in jadwal_senin + jadwal_selasa:
        if len(jadwal_rabu) < batas:
          jadwal_rabu.append(df_jkantor["NAMA_PEGAWAI"][i])
        

#kamis
for i in range(0, len(df_jkantor.NAMA_PEGAWAI)): 
  if df_jkantor["NAMA_PEGAWAI"][i] in bisa_piket.Kamis.values:
    if df_jkantor["NAMA_PEGAWAI"][i] in co:
      if df_jkantor["NAMA_PEGAWAI"][i] not in co_senin + co_selasa + co_rabu:
        if len(co_kamis) < batas_co:
          co_kamis.append(df_jkantor["NAMA_PEGAWAI"][i])
        
    elif df_jkantor["NAMA_PEGAWAI"][i] in kpp.values:
      if df_jkantor["NAMA_PEGAWAI"][i] not in kpp_senin + kpp_selasa + kpp_rabu:
        if len(kpp_kamis) < batas_kpp:
          kpp_kamis.append(df_jkantor["NAMA_PEGAWAI"][i])
        
    elif df_jkantor["NAMA_PEGAWAI"][i] in pel.values:
      if df_jkantor["NAMA_PEGAWAI"][i] not in pel_senin + pel_selasa + pel_rabu:
        if len(pel_kamis) < batas_pel:
          pel_kamis.append(df_jkantor["NAMA_PEGAWAI"][i])
       
    else: 
      if df_jkantor["NAMA_PEGAWAI"][i] not in jadwal_senin + jadwal_selasa + jadwal_rabu:
        if len(jadwal_kamis) < batas:
          jadwal_kamis.append(df_jkantor["NAMA_PEGAWAI"][i])
       

#jumat
for i in range(0, len(df_jkantor.NAMA_PEGAWAI)): 
  if df_jkantor["NAMA_PEGAWAI"][i] in bisa_piket.Jumat.values:
    if df_jkantor["NAMA_PEGAWAI"][i] in co:
      if df_jkantor["NAMA_PEGAWAI"][i] not in co_senin + co_selasa + co_rabu + co_kamis:
        if len(co_jumat) < batas_co:
          co_jumat.append(df_jkantor["NAMA_PEGAWAI"][i])
        
    elif df_jkantor["NAMA_PEGAWAI"][i] in kpp.values:
      if df_jkantor["NAMA_PEGAWAI"][i] not in kpp_senin + kpp_selasa + kpp_rabu + kpp_kamis:
        if len(kpp_jumat) < batas_kpp:
          kpp_jumat.append(df_jkantor["NAMA_PEGAWAI"][i])
       
    elif df_jkantor["NAMA_PEGAWAI"][i] in pel.values:
      if df_jkantor["NAMA_PEGAWAI"][i] not in pel_senin + pel_selasa + pel_rabu + pel_kamis:
        if len(pel_jumat) < batas_pel:
          pel_jumat.append(df_jkantor["NAMA_PEGAWAI"][i])
       
    else: 
      if df_jkantor["NAMA_PEGAWAI"][i] not in jadwal_senin + jadwal_selasa + jadwal_rabu + jadwal_kamis:
        if len(jadwal_jumat) < batas:
          jadwal_jumat.append(df_jkantor["NAMA_PEGAWAI"][i])
        
```

```python
data3 = {'Senin': pd.Series(co_senin + kpp_senin + pel_senin + jadwal_senin),
        'Selasa': pd.Series(co_selasa + kpp_selasa + pel_selasa + jadwal_selasa),
        'Rabu': pd.Series(co_rabu + kpp_rabu + pel_rabu + jadwal_rabu),
        'Kamis': pd.Series(co_kamis + kpp_kamis + pel_kamis + jadwal_kamis),
        'Jumat': pd.Series(co_jumat + kpp_jumat + pel_jumat + jadwal_jumat)}

jadwal_pps = pd.DataFrame(data3)
jadwal_pps
```
